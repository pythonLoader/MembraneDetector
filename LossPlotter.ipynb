```python
import matplotlib.pyplot as plt
import pickle
```

```python
def plot_xyz(train,valid,title = "Loss plot"):
    x = range(len(train))
    plt.plot(train,label="Train")
    plt.plot(valid,label="Validation")
    plt.legend()
    plt.title(title)

    
```

```python
train = pickle.load(open("multires_or_trainingLoss","rb"))
valid = pickle.load(open("multires_or_validLoss","rb"))
plot_xyz(train,valid,title = "Loss Plot (Original MultiresUnet)")
```

```python
train = pickle.load(open("multires_aug_trainingLoss","rb"))
valid = pickle.load(open("multires_aug_validLoss","rb"))
plot_xyz(train,valid,title = "Loss Plot (Augmented MultiresUnet)")
```

```python
train = pickle.load(open("unet_or_trainingLoss","rb"))
valid = pickle.load(open("unet_or_validLoss","rb"))
plot_xyz(train,valid,title = "Loss Plot (Original UNET)")
```

```python
train = pickle.load(open("unet_aug_trainingLoss","rb"))
valid = pickle.load(open("unet_aug_validLoss","rb"))
plot_xyz(train,valid,title = "Loss Plot (Augmented UNET)")
```

```python
def plot_all_train():
    train_or_multires = pickle.load(open("multires_or_trainingLoss","rb"))
    train_aug_multires = pickle.load(open("multires_aug_trainingLoss","rb"))
    
    train_or_unet = pickle.load(open("unet_or_trainingLoss","rb"))
    train_aug_unet = pickle.load(open("unet_aug_trainingLoss","rb"))
    
    plt.plot(train_or_multires,label="Multires Original")
    plt.plot(train_aug_multires,label = "Multires Augment")
    plt.plot(train_or_unet,label="Unet Original")
    plt.plot(train_aug_unet,label="Unet Augmented")
    plt.legend()
plot_all_train()
```

```python
def plot_all_valid():
    # train_or_multires = pickle.load(open("multires_or_validLoss","rb"))
    # train_aug_multires = pickle.load(open("multires_aug_validLoss","rb"))
    
    train_or_unet = pickle.load(open("unet_or_validLoss","rb"))
    train_aug_unet = pickle.load(open("unet_aug_validLoss","rb"))
    
    # plt.plot(train_or_multires,label="Multires Original")
    # plt.plot(train_aug_multires,label = "Multires Augment")
    plt.plot(train_aug_unet,label="Unet Non-Augmented")
    plt.plot(train_or_unet,label="Unet Augmented")
    
    plt.title("Validation Loss")
    plt.xlabel("Epochs")
    plt.xticks(range(0,20))
    plt.legend()
plot_all_valid()
```
